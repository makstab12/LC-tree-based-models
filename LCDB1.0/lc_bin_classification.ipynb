### kr-vs-kp (ID: 3) - JEST

```python
curve1 = lcdb.get_curve("kr-vs-kp", "sklearn.tree.DecisionTreeClassifier")
curve2 = lcdb.get_curve("kr-vs-kp", "sklearn.ensemble.RandomForestClassifier")
```

```python
fig, ax = plt.subplots(1, 2, figsize=(20, 4))
lcdb.plot_train_and_test_curve(curve1, ax[0])
ax[0].set_ylabel("Decision Tree")
lcdb.plot_train_and_test_curve(curve2, ax[1])
ax[1].set_ylabel("Random Forest")
plt.show()
```
